```python
# Calculator Program

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+ - * /): ")

if op == "+":
    result = a + b
    print("Answer is:", result)

elif op == "-":
    result = a - b
    print("Answer is:", result)

elif op == "*":
    result = a * b
    print("Answer is:", result)

elif op == "/":
    result = a / b
    print("Answer is:", result)

else:
    print("Wrong operator")
```

```python
# To Do List

tasks = []

while True:
    print("1 Add Task")
    print("2 Show Task")
    print("3 Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        t = input("Enter task: ")
        tasks.append(t)

    elif choice == "2":
        for i in tasks:
            print(i)

    elif choice == "3":
        break

    else:
        print("Wrong choice")
```

```python
board = ["1","2","3","4","5","6","7","8","9"]
player = "X"

for i in range(9):

    print(board[0], board[1], board[2])
    print(board[3], board[4], board[5])
    print(board[6], board[7], board[8])

    pos = int(input("Enter position (1-9): ")) - 1

    if board[pos] != "X" and board[pos] != "O":
        board[pos] = player
    else:
        print("Already taken")
        continue

    # Simple win check
    if board[0]==board[1]==board[2] or \
       board[3]==board[4]==board[5] or \
       board[6]==board[7]==board[8]:
        print("Player", player, "wins")
        break

    if player == "X":
        player = "O"
    else:
        player = "X"
```

```python
word = "apple"
guess_word = ""
chances = 5

while chances > 0:

    display = ""

    for letter in word:
        if letter in guess_word:
            display = display + letter
        else:
            display = display + "_"

    print("Word:", display)

    if display == word:
        print("You win")
        break

    letter = input("Enter letter: ")
    guess_word = guess_word + letter

    if letter not in word:
        chances = chances - 1
        print("Wrong guess")
        print("Chances left:", chances)

if chances == 0:
    print("You lose")
```
